## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

**Task3**:

Plot the following distributions: $Beta(1,1)$, $Beta(0.9,0.9)$, $Beta(2,1)$, $Beta(1,2)$, $Beta(7,11)$, $Beta(96,42)$, $Beta(5,5)$, $Beta(500,500)$, $Beta(0.8,0.9)$.

**Task10**:

Describe what process each of the previous discrete distributions models, and their support.

Hint: Make sure to read any of the preliZ documentation, the wikipedia articles, or wolfram alpha.


**Task 10 (answer):**

Bernoulli: supports x values between 0 and 1, models the probability of successes (represented at x=1 as a value between 0 and 1) and failures (represented at x=0 as a value between 0 and 1) (Preliz).

Binomial: supports x values from 0 to n (infinity), models the discrete probability distribution of the number of successes in a sequence of n independent yes/no experiments, each of which yields success with probability p (Preliz).

Poisson: supports whole number x values (0 included), models the number of events occurring in a fixed period of time when the times at which events occur are independent (Preliz).

Geometric: supports natural number x values (not 0), models the probability that the first success in a sequence of Bernoulli trials occurs on the xth trial (Preliz).

HyperGeometric: supports x values from 0 to min(k,n) (infinity), models the probability of x successes in a sequence of n bernoulli trials taken without replacement  from a population of N objects, containing k "good" objects (Preliz).

Categorical: supports non-negative integer x values up until length(p)-1 (infinity), models the p values on the y-axis with one y value per x integer value (Preliz).
